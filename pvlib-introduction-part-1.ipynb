## Step 2: Example of using pvlib

Here's a simple example of running pvlib code: using pvlib-python and pandas (a library for data analysis), calculate a clear-sky irradiance profile for Monterrey, Mexico, for today (February 18th, 2026).

```python
# Import libraries
import pvlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define location object
location = pvlib.location.Location(latitude=25.65, longitude=-100.28)
```

```python
# Define time stamps
times = pd.date_range(
    start='2026-02-18 00:00',
    end='2026-02-18 23:59',
    freq='5min',
    tz='America/Monterrey',
)

times
```

```python
# Estimate clear-sky irradiance
clearsky_irradiance = location.get_clearsky(times)
```

```python
# Plot clear-sky irradiance
clearsky_irradiance.plot(ylabel='Irradiance [W/m2]', title='Clear-Sky Simulation')
```
